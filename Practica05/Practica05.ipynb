8. Limpieza de Datos.

Dependiendo del objeto del análisis algunas columnas o registro no serán necesarias en nuestro Dataframe, en este caso omitiremos los datos vacios, por los que eliminaremos la columna 'Type 2', y renombraremos la columna 'Type 1' por 'Type'.

```python
pokemonsE.info()
```

```python
del pokemons['Type 2']
pokemons.rename(columns={'Type 1':'Type'}, inplace= True)
```

```python
del pokemonsE['Type 2']
pokemonsE.rename(columns={'Type 1':'Type'}, inplace= True)
```

11. Calcular el total de datos no duplicados(único)

Usaremos **.len()** como función para calcular la longitud del dataframe y **.unique()** para excluir los duplicados
    
***Nota:*** No considerar ID como un atributo de llave primaria, sino el total de columnas como un solo elemento u objeto.

Obteniendo los valores unicos por columna en este caso en la columna "Name"

```python
len(pokemons.Name.unique())
```

Obteniendo el total de objetos (conjunto de atributos, estrictamente iguales) únicos

```python
total_unicos = len(pokemons.drop_duplicates())
total_unicos
```
