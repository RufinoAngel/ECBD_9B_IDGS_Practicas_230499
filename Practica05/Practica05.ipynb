15. Conteo de pokemones por el atribute Tipo (Graficado)

Se usa la libreria matplot para configurar el tamaño de la gráfica.

NOTA: Por el número de pokemones de tipo agua y normales alto comparado al resto, los pokemones respectivos tendran siempre un total de stats mas alto.

```python
plt.figure(figsize=(10, 5))

ax = sns.countplot(
    data=pokemons,
    x='Type',
    hue='Type',
    palette='twilight',
    legend=False
)

# Agregar totales en la parte superior de cada barra
for container in ax.containers:
    ax.bar_label(
        container,
        label_type='edge',
        padding=3
    )

plt.title('Número de Pokémones agrupados por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Ejercicio 09: Obtener el grafico de total de pokémones por tipo en el DF modificado. Ordenados alfabéticamente

```python
plt.figure(figsize=(10, 5))

# Generamos el orden alfabético limpio de nulos
orden_alfabetico = sorted(pokemons['Type'].dropna().unique())

ax = sns.countplot(
    data=pokemons,
    x='Type',
    hue='Type',
    order=orden_alfabetico,  # <--- Agregamos esto para mantener el orden alfabético
    palette='twilight',
    legend=False
)

# Agregar totales en la parte superior de cada barra
for container in ax.containers:
    ax.bar_label(
        container,
        label_type='edge',
        padding=3
    )

plt.title('Número de Pokémones agrupados por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

22. Calculando los datos par graficción del total de estadisticas por cada tipo de Pokemon, respetando el orden alfabetico previamente definido

```python
plt.style.use('seaborn-v0_8-darkgrid')

stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
palettes = ['magma', 'ocean', 'vlag', 'copper', 'mako', 'winter'] 
pokemon_stats = pokemons.groupby('Type')[stats].sum()
list_types_valida = [t for t in list_types if t in pokemon_stats.index]
pokemon_stats = pokemon_stats.reindex(list_types_valida)
plt.figure(figsize=(17, 17))
for k, stat in enumerate(stats, start=1):
    plt.subplot(3, 2, k)
    sns.barplot(
        x=pokemon_stats[stat],
        y=pokemon_stats.index,
        hue=pokemon_stats.index, # evita warning de palette
        palette=palettes[k-1],
        legend=False
    )

    plt.title(f'Total of {stat}')
    plt.xlabel(stat)
plt.tight_layout()
plt.show()
```

**Ejercicio 13:** Recalcular los gráficos de estadías para el segundo dataframe.

```python
plt.style.use('seaborn-v0_8-darkgrid')

stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
palettes = ['viridis', 'plasma', 'inferno', 'cool', 'spring', 'hot']
pokemon_stats = pokemonsE.groupby('Type')[stats].sum()
tipos_reales_ordenados = sorted(list(pokemon_stats.index))
pokemon_stats = pokemon_stats.reindex(tipos_reales_ordenados)
plt.figure(figsize=(17, 17))
for k, stat in enumerate(stats, start=1):
    plt.subplot(3, 2, k)
    
    sns.barplot(
        x=pokemon_stats[stat],
        y=pokemon_stats.index,
        hue=pokemon_stats.index,
        palette=palettes[k-1],
        legend=False
    )
    
    plt.title(f'Total of {stat}')
    plt.xlabel(stat)
    plt.ylabel('Type')

plt.tight_layout()
plt.show()
plt.figure(figsize=(16, 20))
```
